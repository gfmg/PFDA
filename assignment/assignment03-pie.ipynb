```python
# Import libraries we will use
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Load the data
url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download'
df = pd.read_csv(url,parse_dates=['Date of birth'])
```

```python
df.head()
```

```python
df.info()
```

```python
# Extract email domains from the raw email 
df['domain'] = df['Email'].str.split('@').str[1]

#Group and quantify number of email domains
domain_count = df['domain'].value_counts()
#column names
res=pd.DataFrame(domain_count)
res.columns=['count']
```

```python
# Making a piechart
fig, ax = plt.subplots(figsize=(8,8)) 

ax.pie(res['count'], labels=res.index, autopct='%1.2f%%', startangle=90,shadow=True,
       colors=sns.color_palette('muted'),textprops={'fontsize': 14})
plt.title("Email domains percentages", fontweight="bold",fontdict={'fontsize': 20})
```
